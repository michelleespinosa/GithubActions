```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define function for plotting points
def plot_points(ax, points, title, color='blue'):
    ax.scatter(points[0, :], points[1, :], color=color)
    # Annotating each point with labels 1 to 6
    for i in range(points.shape[1]):
        ax.annotate(f"{i+1}", (points[0, i], points[1, i]), fontsize=10,
                    verticalalignment='bottom', horizontalalignment='right')
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.axhline(0, color='gray', linewidth=0.5)
    ax.axvline(0, color='gray', linewidth=0.5)
    ax.grid(True, linestyle='--', linewidth=0.5)
    ax.set_title(title)
```

```python
# Define the original points, including (0,0), (0,1), (1,0), (1,1), (0,2), and (1,2)
points = np.array([[0, 0, 0, 1, 1, 1], [0, 1, 2, 0, 1, 2]])  # 2x6 matrix

# Plot original points
fig, ax = plt.subplots(figsize=(5, 5))
plot_points(ax, points, "Original Points")
plt.show()
```

```python
# Define the rotation matrix for theta (pi/4 radians)
theta = np.pi / 4
A = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])  #Rotation by pi/4 radians
```

```python
# Apply the transformation A to the points
transformed_points_A = A @ points #  you can also @ for matrix multiplication (recommended in Python 3.5+)
# altneratively you could use np.dot(A,points)
```

```python
# plot original points and transformed points
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
plot_points(axes[0], points, "Original Points x")
plot_points(axes[1], transformed_points_A, "Transformed points Ax ", color='red')
plt.show()
```

***
What if we apply $A^{-1}$?


```python
# compute the inverse
B = np.linalg.inv(A) # B = A^{-1}

# Apply B to the transformed points Ax
transformed_points_BA = B @ transformed_points_A  # Applying B to the transformed points
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot the transformed points by A
plot_points(axes[0], transformed_points_A, "Transformed points  Ax ", color='red')

# Plot the points after applying B (inverse of A)
plot_points(axes[1], transformed_points_BA, "Transformed points B(Ax) ", color='green')

plt.show()
```

***
Let's see what B does to the original points.

```python
# Apply B= A^{-1} to the orginal points
transformed_points_B = B @ points  # Applying B to the original points

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot the transformed points by A
plot_points(axes[0], points, "Original points x ", color='blue')

# Plot the points after applying B (inverse of A)
plot_points(axes[1], transformed_points_B, "Transformed points Bx ", color='black')

plt.show()
```
